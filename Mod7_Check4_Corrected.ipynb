```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Data/IMB_HR.csv')
>>> print(df)
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.boxplot(df['Age'])
plt.title('Employee Age')
          
plt.subplot(2, 2, 2)
plt.scatter(x = df['Age'], y = df['DailyRate'], color='red')
plt.ylabel('Daily Rate ($)')
plt.xlabel('Age (years)')
plt.title('Age vs Pay')

plt.subplot(2, 2, 3)
plt.scatter(x = df['YearsAtCompany'], y = df['DailyRate'], color='blue')
plt.ylabel('Daily Rate ($)')
plt.xlabel('Total Years Working for IBM (years)')
plt.title('Years Worked vs Pay')

plt.subplot(2, 2, 4)
plt.hist(df['YearsAtCompany'], bins=20)
plt.title('Random Cube Root Distribution')
plt.ylabel('Frequency')
plt.xlabel('Years Worked for IBM')
plt.show
```
