```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame()

df['rand'] = np.random.normal(50, 20, 100)
df['rand_cube_rt'] = df['rand'] ** (1/3)

for x in df['rand']:  #remove the occasional negative outlier
    if x < 0:
        x *= -1
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.boxplot(df['rand'])

plt.subplot(2, 2, 2)
plt.scatter(x = df['rand_cube_rt'], y = df['rand'], color='red')
plt.ylabel('Values')
plt.xlabel('Cube Roots')
plt.title('Random Values Cube Rooted')

plt.subplot(2, 2, 3)
plt.plot(df['rand_cube_rt'])
plt.ylabel('Cube Roots')
plt.xlabel('Index')
plt.title('Random Values Cube Rooted')

plt.subplot(2, 2, 4)
plt.hist(df['rand_cube_rt'], bins=20)
plt.title('Random Cube Root Distribution')
plt.ylabel('Frequency')
plt.xlabel('Cube Roots')
plt.show
```
