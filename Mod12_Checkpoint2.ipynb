```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
sns.set()
```

```python
#assign data frames
df = pd.read_csv('Data/IMB_HR.csv')
df_quitters = df.loc[lambda df: df['Attrition'] == 'Yes', :]
df_stayers = df.loc[lambda df: df['Attrition'] == 'No', :]
```

## Choose one variable and plot that variable four different ways.

```python
sns.distplot(df_quitters['MonthlyIncome']);
plt.show()

sns.distplot(df_quitters['MonthlyIncome'], kde=False, rug=True, color='purple');
plt.show()

sns.distplot(df_quitters['MonthlyIncome'], hist=False, rug=True, color='red');
plt.show()

sns.boxplot(df_quitters['MonthlyIncome'], palette="BuGn_r")
plt.show()
```

**Fig. 1**
    *Pros:* Legible, easy to see the distribution;
    *Cons:* Frequency scale seems esoteric
 
**Fig. 2**
    *Pros:* Includes ticks for discrete data points; Numerical scale on y-axis perhaps more useful than frequency;
    *Cons:* Overlap of bins and ticks affects legibility
    
**Fig. 3**
    *Pros:* Legible, easy to see the distribution; Includes ticks for discrete data points;
    *Cons:* Perhaps too general
    
**Fig. 4**
    *Pros:* Boxplot shows a lot of information in a concise format;
    *Cons:* Maybe not easily understood by laypersons

## Choose two continuous variables, and plot them three different ways

```python
sns.jointplot(x=df_quitters['MonthlyIncome'], y=df_quitters['Age'], kind="kde")
plt.show()

sns.jointplot(x=df_quitters['MonthlyIncome'], y=df_quitters['Age'], color='purple')
plt.show()

sns.jointplot(x=df_quitters['MonthlyIncome'], y=df_quitters['Age'], kind="hex", color="k")
plt.show()
```

**Fig. 1**
    *Pros:* Visually arresting;
    *Cons:* Too general; some information loss
 
**Fig. 2**
    *Pros:* Attractive, lots of info;
    *Cons:* Areas with a lot of data points are a bit muddy
    
**Fig. 3**
    *Pros:* Attractive, informative, highlights relative frequency well;
    *Cons:* n/a

## Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters)
plt.show()

sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters, kind='bar')
plt.show()

sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters, kind='swarm')
plt.show()

sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters, kind='box')
plt.show()

sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters, kind='violin')
plt.show()

sns.catplot(x='MonthlyIncome', y='Department', data=df_quitters, kind='violin', inner='stick', palette="pastel")
plt.show()
```

**Fig. 1**
    *Pros:* Legible, easy to see clusters and outliers;
    *Cons:* Maybe not the most attractive
 
**Fig. 2**
    *Pros:* Simple; includes error bars;
    *Cons:* No visibility of discrete data
    
**Fig. 3**
    *Pros:* Legible, even better than fig. 1 for clusters;
    *Cons:* Not the most attractive
    
**Fig. 4**
    *Pros:* Lots of info;
    *Cons:* Maybe not easily understood by laypersons
    
**Fig. 5**
    *Pros:* Shows the distribution in a more attractive way;
    *Cons:* Discrete data is kind of lost
    
**Fig. 6**
    *Pros:* Attractive, shows discrete data points;
    *Cons:* Perhaps the shapes could be confusing to laypersons
    
    

## Find a dataset with at least four continuous variables and one categorical variable. Create one master plot that gives insight into the variables and their interrelationships

```python
sns.relplot(x="Age", y="MonthlyIncome", hue="Department", data=df_quitters)
plt.show()

sns.relplot(x="MonthlyIncome", y="MonthlyRate", hue="Department", data=df_quitters)
plt.show()
```

The figure shows that HR employees tend to get paid less than the other departments.  There is a conspicuous vertical line of R&D employees at around 2500 dollars of MonthlyIncome that extends beyond 25000 dollars of MonthlyRate - perhaps these employees were receiving constly training services.  Unfortunately, no other obvious trends emerge from these figures.
